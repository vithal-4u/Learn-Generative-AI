```python
corpus="""
Arjuna's Charioteer and Guide: Krishna chose to be Arjuna's charioteer, a seemingly humble role that enabled him to be close to Arjuna and offer constant guidance throughout the war.
The Bhagavad Gita: As Arjuna faced a profound moral dilemma before the war, hesitant to fight his own family members and teachers, Krishna delivered the teachings of the Bhagavad Gita. This seminal text emphasizes concepts of dharma (duty), karma (action), and moksha (liberation), urging Arjuna to fulfill his duty as a warrior without being attached to the results.
A Non-Combative Role (with a twist): Krishna initially promised not to wield weapons in the battle. However, he still played an active role through strategic advice, diplomacy, and subtle interventions to ensure the victory of righteousness (dharma).
"""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
