# **BOW Example**

```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
```

### **Step-1: Preprocessing the Text**

Before applying the BoW model, we need to preprocess the text. This includes:  
    ▪ Converting the text to lowercase  
    ▪ Removing non-word characters.  
    ▪ Removing extra spaces.  

```python
text = """Beans. I was trying to explain to somebody as we were flying in, that's corn.  
    That's beans. And they were very impressed at my agricultural knowledge. Please give it up for Amaury once again for that 
    outstanding introduction. I have a bunch of good friends here today, including somebody who I served with, 
    who is one of the finest senators in the country, and we're lucky to have him, your Senator, Dick Durbin is here. 
    I also noticed, by the way, former Governor Edgar here, who I haven't seen in a long time, and somehow he has not aged and I have. 
    And it's great to see you, Governor. I want to thank President Killeen and everybody at the U of I System for making 
    it possible for me to be here today. And I am deeply honored at the Paul Douglas Award that is being given to me. 
    He is somebody who set the path for so much outstanding public service here in Illinois. 
    Now, I want to start by addressing the elephant in the room. I know people are still wondering why I didn't speak at the commencement."""

dataset = nltk.sent_tokenize(text)
```

```python
for i, sentence in enumerate(dataset):
    print(f"Sentence {i+1}: {sentence}")
```

```python
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W', ' ', dataset[i])
    dataset[i] = re.sub(r'\s+', ' ', dataset[i])
```

```python
for i, sentence in enumerate(dataset):
    print(f"Sentence {i+1}: {sentence}")
```

### **Step-2: Counting Word Frequencies**. 
We count the frequency of each word in the preprocessed text. We will store these counts in a pandas DataFrame to view them easily in a tabular format.  
    ▪ We initialize a dictionary to hold our word counts.  
    ▪ Then, we tokenize each sentence into words.  
    ▪ For each word, we check if it exists in our dictionary. If it does, we increment its count. If it doesn’t, we add it to the dictionary with a count of 1.  

```python
word2count = {}

for data in dataset:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count:
            word2count[word] = 1
        else:
            word2count[word] += 1
```

```python
word2count
```

```python
stop_words = set(stopwords.words('english'))

#filtered_word2count = {word: count for word, count in word2count.items() if word not in stop_words}

# Initialize an empty dictionary to store the filtered words and their counts
filtered_word2count = {}

# Iterate through each word and count in the original dictionary
for word, count in word2count.items():
    #print(f"word {i+1}: {word}  count: {count}")
    # Check if the current word is not in the stop words list
    if word not in stop_words:
        # If it's not a stop word, add it to our filtered dictionary
        filtered_word2count[word] = count

# Now filtered_word2count contains only words that aren't stop words
```

```python
# DataFrame after removing stopwords:
filtered_word2count
```

```python
word_freq_df = pd.DataFrame(list(filtered_word2count.items()), columns=['Word', 'Frequency'])

word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

print(word_freq_df)
```

### **Step 3: Selecting the Most Frequent Words**. 
Now that we have counted the word frequencies, we will select the top N most frequent words (e.g top 10) to be used in the BoW model. We can visualize these frequent words using a bar chart to understand the distribution of words in our dataset.

```python
import heapq
import matplotlib.pyplot as plt

freq_words = heapq.nlargest(10, word2count, key=word2count.get)

print(f"Top 10 frequent words: {freq_words}")

#top_words = sorted(word2count.items(), key=lambda x: x[1], reverse=True)[:10]

# Step 1: Convert the dictionary items to a list of tuples
word_count_pairs = list(word2count.items())

# Step 2: Define a function to extract the count (second element) from each tuple
def get_count(item):
    #print(f"item : {item[1]}")
    return item[1]

# Step 3: Sort the list of tuples based on the count
# The reverse=True parameter sorts in descending order
sorted_word_counts = sorted(word_count_pairs, key=get_count, reverse=True)

# Step 4: Take only the first 10 items from the sorted list
top_words = sorted_word_counts[:10]

# Now top_words contains the 10 most frequent words with their counts

words, counts = zip(*top_words)

plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Most Frequent Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

### **Step 4: Building the Bag of Words (BoW) Model**. 
Now we will build the Bag of Words (BoW) model. This model is represented as a binary matrix where each row corresponds to a sentence and each column represents one of the top N frequent words. A 1 in the matrix shows that the word is present in the sentence and a 0 shows its absence.
We will use a heatmap to visualize this binary matrix where green shows the presence of a word (1) and red shows its absence (0).

```python

import numpy as np
import seaborn as sns

X = []
for data in dataset:
    vector = []
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)

X = np.asarray(X)
for i, sentence in enumerate(dataset):
    print(f"Sentence {i+1}: {sentence}")


sizeAr = len(dataset)
plt.figure(figsize=(sizeAr, sizeAr))
sns.heatmap(X, cmap='RdYlGn', cbar=False, annot=True, fmt="d", xticklabels=freq_words, yticklabels=[f"Sentence {i+1}" for i in range(len(dataset))])

plt.title('Bag of Words Matrix')
plt.xlabel('Frequent Words')
plt.ylabel('Sentences')
plt.show()
```

### **Step 5: Visualizing Word Frequencies with a Word Cloud**

we can create a Word Cloud to visually represent the word frequencies. In a word cloud, the size of each word is proportional to its frequency which makes it easy to identify the most common words at a glance.

```python

from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word2count)
plt.figure(figsize=(sizeAr, sizeAr))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Frequent Words")
plt.show()
```
